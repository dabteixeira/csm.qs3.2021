###Exibição de imagens P&B

### Alterando dimensões da imagem


```python
from google.colab.patches import cv2_imshow # cv2_imshow
img1RGB = cv.cvtColor(imagem1,cv.COLOR_BGR2RGB)
img2 = cv.resize(img1RGB, None, fx=1.25, fy=1.25, interpolation = cv.INTER_CUBIC)
img4 = cv.resize(img1RGB, None, fx=0.50, fy=0.50, interpolation = cv.INTER_CUBIC)

print("50% da Dimensão original")
print("Formato da imagem: ", img4.shape)
print(f"Esta imagem contém {img4.shape[0] * img4.shape[1]} pixels")
cv2_imshow(img4)
print(" ")
print("125% da Dimensão original")
print("Formato da imagem: ", img2.shape)
print(f"Esta imagem contém {img2.shape[0] * img2.shape[1]} pixels")
cv2_imshow(img2)
print(" ")

img2RGB = cv.cvtColor(imagem2,cv.COLOR_BGR2RGB)
img3 = cv.resize(img2RGB, None, fx=1.1, fy=1.1, interpolation = cv.INTER_CUBIC)
img6 = cv.resize(img2RGB, None, fx=0.75, fy=0.75, interpolation = cv.INTER_CUBIC)

print("75% da Dimensão original")
print("Formato da imagem: ", img6.shape)
print(f"Esta imagem contém {img6.shape[0] * img6.shape[1]} pixels")
cv2_imshow(img6)
print(" ")
print("110% da Dimensão original")
print("Formato da imagem: ", img3.shape)
print(f"Esta imagem contém {img3.shape[0] * img3.shape[1]} pixels")
cv2_imshow(img3)
cv.imwrite("/content/drive/My Drive/Colab Notebooks/Colab_Lab2_2021/grupo_50.jpg", img4)
cv.imwrite("/content/drive/My Drive/Colab Notebooks/Colab_Lab2_2021/grupo_125.jpg", img2)
cv.imwrite("/content/drive/My Drive/Colab Notebooks/Colab_Lab2_2021/avatares_75.jpg", img6)
cv.imwrite("/content/drive/My Drive/Colab Notebooks/Colab_Lab2_2021/avatares_110.jpg", img3)
```

### Reprodução de vídeos. Resolução 100%


```python
from IPython.display import HTML
from base64 import b64encode
video_path = "/content/drive/My Drive/Colab Notebooks/Colab_Lab2_2021/Pikachu.mp4"
mp4 = open(video_path,'rb').read()
video_decodificado = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f'<video width=720 controls><source src= {video_decodificado} type="video/mp4"></video>')
```
